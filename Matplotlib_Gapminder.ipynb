```python
# import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# load the data
gapminder_df = pd.read_csv('gapminder_2.csv', index_col = 'country')
```

```python
# display the top 20 records
gapminder_df.head(20)
```

```python
# Make a line plot, gdp_cap on the x-axis, life_exp on the y-axis
plt.plot(gapminder_df['gdp_cap'], gapminder_df['life_exp'])
# Display the plot
plt.show()
```

```python
# the line doesn't look right.
# in case you're trying to assess if there's a correlation between two variables,the scatter plot is the better choice.
# Change the line plot below to a scatter plot
plt.scatter(gapminder_df['gdp_cap'], gapminder_df['life_exp'])
# Put the x-axis on a logarithmic scale
plt.xscale('log')
# Display the plot
plt.show()
```

```python
# scatter plot
plt.scatter(gapminder_df['gdp_cap'], gapminder_df['life_exp'])
# Put the x-axis on a logarithmic scale
plt.xscale('log')
# Add axis labels & Title
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
# Display the plot
plt.show()
```

```python
# scatter plot
plt.scatter(gapminder_df['gdp_cap'], gapminder_df['life_exp'])
# Put the x-axis on a logarithmic scale
plt.xscale('log')
# Add axis labels & Title
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
# Adapt the ticks on the x-axis
plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])
# Display the plot
plt.show()
```

```python
plt.figure(dpi=150)
# Store population as a numpy array: np_pop
np_pop = np.array(gapminder_df['population'])
# Double np_pop array
np_pop2 = np_pop * 2
#Use seaborn scatterplot for better customization
sns.scatterplot(gapminder_df['gdp_cap'], gapminder_df['life_exp'], 
                hue = gapminder_df['cont'], size = np_pop2, sizes=(20,400), legend = False)
# Add grid() call
plt.grid(True)
# Put the x-axis on a logarithmic scale
plt.xscale('log')
# Add axis labels & Title
plt.xlabel('GDP per Capita [in USD]', fontsize = 14)
plt.ylabel('Life Expectancy [in years]', fontsize = 14)
plt.title('World Development in 2007', fontsize = 20)
# Adapt the ticks on the x-axis
plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])
#Add description to the biggest countries
plt.text(1550, 67, 'India')
plt.text(5650, 75, 'China')
# Display the plot
plt.show()
```

```python
# Create histogram of life_exp data
# no of bins is 10 by default
plt.hist(gapminder_df['life_exp'])
# Display histogram
plt.show()
```
